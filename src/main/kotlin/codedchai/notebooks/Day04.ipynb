```kotlin
import codedchai.println
import codedchai.readGridInput

val gridInput = readGridInput("Day04")

fun Char.isPaperRoll() = this == '@'

gridInput.coordinatesToValues.keys.count { coordinate ->
  val cell = gridInput.get(coordinate)!!
  if (cell.isPaperRoll()) {
    coordinate.adjacentNeighbors().count { neighbor ->
      val neighbor = gridInput.get(neighbor)
      neighbor != null && neighbor.isPaperRoll()
    } < 4
  } else {
    false
  }
}.println()
```
