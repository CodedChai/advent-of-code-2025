```kotlin
import codedchai.Grid
import codedchai.Vec2
import codedchai.println
import codedchai.readGridInput

val gridInput = readGridInput("Day04")

fun Char.isPaperRoll() = this == '@'

fun paperTowelsToRemove(grid: Grid<Char>): List<Vec2> {
  return grid.coordinatesToValues.keys.mapNotNull { coordinate ->
    val cell = grid[coordinate]!!
    val canRemove = if (cell.isPaperRoll()) {
      coordinate.adjacentNeighbors().count { neighborCoord ->
        val neighbor = grid[neighborCoord]
        neighbor != null && neighbor.isPaperRoll()
      } < 4
    } else {
      false
    }
    if (canRemove) {
      coordinate
    } else {
      null
    }
  }
}

println("Part 1 answer: ${paperTowelsToRemove(gridInput).size}")

var removedAny = true
var totalRemoved = 0
var currentGrid = gridInput.deepCopy()
while (removedAny) {
  val coordinatesToRemove = paperTowelsToRemove(currentGrid).toHashSet()

  val newGridMap = currentGrid.coordinatesToValues.keys.map { coord ->
    val newValue = if (coord in coordinatesToRemove) {
      '.'
    } else {
      currentGrid[coord]!!
    }
    coord to newValue
  }.toMap(hashMapOf())

  currentGrid = Grid(newGridMap)

  removedAny = coordinatesToRemove.isNotEmpty()
  totalRemoved += coordinatesToRemove.size
  println("Removed: ${coordinatesToRemove.size}")
}

println("Part 2 answer: $totalRemoved")
```
